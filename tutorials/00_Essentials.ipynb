## Install StratiLib

You can find more details in the [github source page of StratiLib](https://github.com/rubensdmp/StratiLib)

```python
# Installs stratilib package
import subprocess

try:
    import stratilib as sl
except ImportError:
    print('stratilib package not installed. Installing ...')
    subprocess.check_call(["python", '-m', 'pip', 'install', 'stratilib'])
```

### SHOW LITHOLOGIES

show_litho() plot the lithologies that StratiLib provides, and also you can give it some parameters:

**sedlog = True** : show only SedLog 3.1 lithologies <br>
**lang = {'sp','es'}** : language sp:spanish, es:english <br>
**df = pandas dataframe** from read_litho or read_sedlog <br>
**rows = number** of rows to plot lithologys <br>
**cols = number** of cols to plot lithologys <br>
**tit = title**, default 'Litologies' <br>
**keys = True** : show lithology keys <br>
**save_fig = name** of file to be saved <br>

```python
#example with keys in brackets
sl.show_litho(tit = 'Lithologies that you can use', keys = True)
```

### SHOW STRUCTURES AND FOSSILS

show_structs() plot the structures and fossils that StratiLib provides, and also you can give it some parameters:

**df = pandas dataframe** from read_litho or read_sedlog <br>
**plot {'structures', 'fossils', 'both'}** default 'both' <br>
**keys = True** : show fossils and structures keys <br>
**save_fig = name** of file to be saved <br>

```python
#example with keys in brackets
sl.show_structs(keys = True)
```
