```python
import stratilib as sl
from google.colab import files
```

## **LOAD FILE**

Click [**Here**](https://github.com/rubensdmp/StratiLib/blob/main/tutorials/data/StratiLib_Template_FULL.xlsx?raw=true) to download the template to be loaded in StratiLib <br>
Then you can charge your data in "DATA" sheet, and select this file running next cell.<br>
Note: You can see the [**DEFINITIONS**](https://github.com/rubensdmp/StratiLib/blob/main/tutorials/data/DEFINITIONS.xlsx?raw=true) sheet for more information


```python
uploaded = files.upload()
for fn in uploaded.keys():
  print('Se cargaron los archivos:"{name}" que pesan: {length} bytes'.format(
      name=fn, length=len(uploaded[fn])))

fn = list(uploaded.keys())[0]
fn
```

## **READ EXCEL FILE WITH STRATILIB FORMAT OR SEDLOG CSV FILE**
Choose which cell run according to the type of file did you loaded

```python
# FOR EXCEL FILE
df = sl.read_litho(fn)
```

```python
# FOR CSV SEDLOG 3.1 FILE
df = sl.read_sedlog(fn)
```

## **PLOT COLUMN**

```python
sl.plot_litho(df,
              top = 0,
              base = 300,
              width = 4,
              length = 4,
              show_structs = True,
              struct_out = True,
              show_fossils = True,
              fossil_out = False,
              show_fm = False,
              only_first_lith =  True,
              show_des = True,
              save_fig='YOUR_FILE_NAME'
              )
```

#### Download plot in png

```python
files.download('YOUR_FILE_NAME.png')
```

## **PLOT LITHOLOGIES OF THE COLUMN**

```python
sl.show_litho(df =df, 
              cols = 4,
              tit = '',
              save_fig='my_file_litho')
```

#### Download lithologies in png

```python
files.download('my_file_litho.png')
```

## **PLOT STRUCTURES AND FOSSILS OF THE COLUMN**

```python
sl.show_structs(df =df,
                ploth = 'both',
                tit = 'strutures and fossils for...',
                save_fig = 'my_file_structs')
```

#### Download structures and fossils in png

```python
files.download('my_file_structs.png')
```

## **EXPORT TO CSV SEDLOG 3.1 FILE**

```python
sl.to_sedlog_csv(df,
                 'exported_file')
```

#### Download file to be imported from SedLog 3.1

```python
files.download('exported_file.csv')
```
